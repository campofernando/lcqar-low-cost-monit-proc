## Upload Results

```python
def rename_sensors_names(old_names):
    names = [name.strip() for name in old_names.split(',')]
    new_names = ""
    for name in names:
        if 'measuring' in name: 
            if len(new_names) > 0: new_names += " + "
            new_name = name.replace('measuring ', '')
            new_names += new_name
        elif name == 'temperature': 
            if len(new_names) > 0: new_names += " + "
            new_names += 'T'
    return new_names
```

## Performance metrics

### R2

#### CO

```python
import ast

r2_dataframe = pd.DataFrame(columns=['Sensores', 'MLP', 'MLR', 'KNN', 'RF'])
r2_original_dataframe = co_r2_dataframe
r2_original_dataframe = r2_original_dataframe.reset_index().drop(columns='index')
r2_dataframe['Sensores'] = r2_original_dataframe['Sensores']
mean_r2_dataframe = r2_dataframe.copy()
mask_dataframe = r2_dataframe.copy()
for index, row in r2_original_dataframe.iterrows():
    row_MLP = ast.literal_eval(row['MLP'])
    row_MLR = ast.literal_eval(row['MLR'])
    row_KNN = ast.literal_eval(row['KNN'])
    row_RF = ast.literal_eval(row['RF'])
    r2_dataframe.loc[index, 'MLP'] = f"{round(row_MLP['Mean'], 2)} Â± {round(row_MLP['Std'], 2)}"
    r2_dataframe.loc[index, 'MLR'] = f"{round(row_MLR['Mean'], 2)} Â± {round(row_MLR['Std'], 2)}"
    r2_dataframe.loc[index, 'KNN'] = f"{round(row_KNN['Mean'], 2)} Â± {round(row_KNN['Std'], 2)}"
    r2_dataframe.loc[index, 'RF'] = f"{round(row_RF['Mean'], 2)} Â± {round(row_RF['Std'], 2)}"
    mean_r2_dataframe.loc[index, 'MLP'] = round(row_MLP['Mean'], 2)
    mean_r2_dataframe.loc[index, 'MLR'] = round(row_MLR['Mean'], 2)
    mean_r2_dataframe.loc[index, 'KNN'] = round(row_KNN['Mean'], 2)
    mean_r2_dataframe.loc[index, 'RF'] = round(row_RF['Mean'], 2)

r2_dataframe.to_csv(output_directory_path + sensor_o3_name + "_r2_formated_results.csv")
r2_dataframe
```

```python
mean_r2_dataframe
positive_r2_mlp = mean_r2_dataframe[mean_r2_dataframe['MLP'] > -1]['MLP']
positive_r2_mlr = mean_r2_dataframe[mean_r2_dataframe['MLP'] > -1]['MLR']
positive_r2_knn = mean_r2_dataframe[mean_r2_dataframe['MLP'] > -1]['KNN']
positive_r2_rf = mean_r2_dataframe[mean_r2_dataframe['MLP'] > -1]['RF']

positive_r2_dataframe = pd.concat([positive_r2_mlp, positive_r2_mlr, positive_r2_knn, positive_r2_rf], axis=1)
positive_r2_dataframe['Sensores'] = mean_r2_dataframe.loc[positive_r2_dataframe.index, 'Sensores']
positive_r2_dataframe.head(9)
```

```python
import numpy as np
import matplotlib.pyplot as plt

y_pos = np.arange(len(positive_r2_dataframe['Sensores']))
plt.figure(figsize=(10,20))

ax_r2  = plt.axes()
ax_r2.scatter(positive_r2_dataframe['MLP'], y_pos)
ax_r2.scatter(positive_r2_dataframe['MLR'], y_pos)
ax_r2.scatter(positive_r2_dataframe['KNN'], y_pos)
ax_r2.scatter(positive_r2_dataframe['RF'], y_pos)
ax_r2.yaxis.grid(True, linestyle='--', color='gray', alpha=0.5)
ax_r2.set_yticks(y_pos, labels=positive_r2_dataframe['Sensores'])
ax_r2.invert_yaxis()  # labels read top-to-bottom

for i, value in enumerate(positive_r2_dataframe['MLP']):
    if not np.isnan(value):
        ax_r2.text(value, i, 'MLP', fontsize=10, ha='right', va='bottom')

for i, value in enumerate(positive_r2_dataframe['MLR']):
    if not np.isnan(value):
        ax_r2.text(value, i, 'MLR', fontsize=10, ha='right', va='bottom')

for i, value in enumerate(positive_r2_dataframe['KNN']):
    if not np.isnan(value):
        ax_r2.text(value, y_pos[i], 'KNN', fontsize=10, ha='right', va='bottom')

for i, value in enumerate(positive_r2_dataframe['RF']):
    if not np.isnan(value):
        ax_r2.text(value, y_pos[i], 'RF', fontsize=10, ha='right', va='bottom')

for i, sensor_index in enumerate(positive_r2_dataframe.index):
    values = list(positive_r2_dataframe.drop(columns='Sensores').iloc[i])
    for j, value in enumerate(values):
        if j < len(values) - 1 and not np.isnan(value) and not np.isnan(values[j + 1]):
            point_a = (value, i)
            point_b = (values[j + 1], i)
            ax_r2.plot([point_a[0], point_b[0]], [point_a[1], point_b[1]], color='black', linestyle='-')


plt.show()
```
