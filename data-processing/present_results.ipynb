## Upload Results

```python
def rename_sensors_names(new_name, old_names):
    names = [name.strip() for name in old_names.split('|')]
    new_names = ""
    for name in names:
        if name == 'measuring': 
            if len(new_names) > 0: new_names += " + "
            new_names += new_name
        if name == 'measuring 1': 
            if len(new_names) > 0: new_names += " + "
            new_names += (new_name + " \N{SUPERSCRIPT ONE}")
        if name == 'measuring 2': 
            if len(new_names) > 0: new_names += " + "
            new_names += (new_name + ' \N{SUPERSCRIPT TWO}')
        elif name == 'temperature': 
            if len(new_names) > 0: new_names += " + "
            new_names += 'T'
    return new_names
```

## Performance metrics

### R2

```python
import ast

r2_dataframe = pd.DataFrame(columns=['Sensores', 'MLP', 'MLR', 'KNN', 'RF'])
r2_original_dataframe = pd.concat([co_r2_dataframe, no2_r2_dataframe, o3_r2_dataframe, so2_r2_dataframe], axis=0)
r2_original_dataframe = r2_original_dataframe.reset_index().drop(columns='index')
r2_dataframe['Sensores'] = r2_original_dataframe['Sensores']
mask_dataframe = r2_dataframe.copy()
for index, row in r2_original_dataframe.iterrows():
    row_MLP = ast.literal_eval(row['MLP'])
    row_MLR = ast.literal_eval(row['MLR'])
    row_KNN = ast.literal_eval(row['KNN'])
    row_RF = ast.literal_eval(row['RF'])
    r2_dataframe.loc[index, 'MLP'] = f"{round(row_MLP['Mean'], 2)} ± {round(row_MLP['Std'], 2)}"
    r2_dataframe.loc[index, 'MLR'] = f"{round(row_MLR['Mean'], 2)} ± {round(row_MLR['Std'], 2)}"
    r2_dataframe.loc[index, 'KNN'] = f"{round(row_KNN['Mean'], 2)} ± {round(row_KNN['Std'], 2)}"
    r2_dataframe.loc[index, 'RF'] = f"{round(row_RF['Mean'], 2)} ± {round(row_RF['Std'], 2)}"
    
r2_dataframe
```
